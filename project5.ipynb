5. LINE, PLANE, AND HYPERPLANE

LINE (1D)

A line represents movement in one direction.

Example:

[80,80,80] + t[1,1,1]

This means scores increase equally in all subjects.

------------------------------------------------------------

PLANE (2D)

A plane is formed using two independent vectors.

Example:

a[1,0,1] + b[0,1,1]

This represents relationships among two changing subject dimensions.

------------------------------------------------------------

HYPERPLANE (HIGHER DIMENSION)

A hyperplane separates data into categories.

Example:

w1x1 + w2x2 + w3x3 + b = 0

In machine learning, hyperplanes are used for classification problems.

PART A : VECTOR & MATRIX FUNDAMENTALS

```python

import numpy as np
from numpy.linalg import norm

# Student Vectors
s1 = np.array([85, 78, 90])
s2 = np.array([88, 76, 85])

print("Student 1 Vector =", s1)
print("Student 2 Vector =", s2)

# Norm-1 and Norm-2

norm1 = norm(s1, 1)
norm2 = norm(s1, 2)

print("\nNorm-1 of S1 =", norm1)
print("Norm-2 of S1 =", norm2)


# Dot Product


dot_product = np.dot(s1, s2)

print("\nDot Product =", dot_product)

# Angle Between Two Vectors

angle = np.arccos(dot_product / (norm(s1) * norm(s2)))
angle_degree = np.degrees(angle)

print("\nAngle Between Vectors =", angle_degree)

# Cross Product

cross_product = np.cross(s1, s2)

print("\nCross Product =", cross_product)


# Projection of One Vector onto Another


projection = (np.dot(s1, s2) / np.dot(s2, s2)) * s2

print("\nProjection of S1 onto S2 =")
print(projection)
```

PART B : MATRIX OPERATIONS

```python

import numpy as np
import pandas as pd
from numpy.linalg import det, inv

data = {
    "Math": [85, 88, 92, 70, 95],
    "Science": [78, 76, 89, 65, 91],
    "English": [90, 85, 94, 72, 96]
}

dataset = pd.DataFrame(data)

# Matrix Formation
A = dataset.values

print("Matrix A:")
print(A)

# Matrix Addition

B = np.ones((5, 3))

addition = A + B

print("\nMatrix Addition (A + B):")
print(addition)

# Matrix Multiplication

multiplication = np.dot(A.T, A)

print("\nMatrix Multiplication (Aᵀ × A):")
print(multiplication)


# Transpose Matrix


transpose = A.T

print("\nTranspose Matrix:")
print(transpose)

# Determinant


M = A[:3]

determinant = det(M)

print("\nDeterminant =", determinant)

# Inverse Matrix

inverse_matrix = inv(M)

print("\nInverse Matrix:")
print(inverse_matrix)
```

PART D : EIGENVALUES & DECOMPOSITION

```python
import numpy as np
import pandas as pd
from numpy.linalg import eig, svd
from scipy.linalg import lu

data = {
    "Math": [85, 88, 92, 70, 95],
    "Science": [78, 76, 89, 65, 91],
    "English": [90, 85, 94, 72, 96]
}

dataset = pd.DataFrame(data)

A = dataset.values

cov_matrix = np.cov(A.T)

print("Covariance Matrix:")
print(cov_matrix)

eigenvalues, eigenvectors = eig(cov_matrix)

print("\nEigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)

M = A[:3]

P, L, U = lu(M)

print("\nLower Triangular Matrix (L):")
print(L)

print("\nUpper Triangular Matrix (U):")
print(U)

U_svd, S_svd, VT_svd = svd(M)

print("\nSingular Values:")
print(S_svd)

print("\nU Matrix:")
print(U_svd)

print("\nVT Matrix:")
print(VT_svd)
```

PART E : DIMENSIONALITY REDUCTION

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler

data = {
    "Math": [85, 88, 92, 70, 95],
    "Science": [78, 76, 89, 65, 91],
    "English": [90, 85, 94, 72, 96]
}

dataset = pd.DataFrame(data)

A = dataset.values

scaler = StandardScaler()

scaled_data = scaler.fit_transform(A)

pca = PCA(n_components=2)

pca_result = pca.fit_transform(scaled_data)

print("PCA Reduced Data:")
print(pca_result)

average_scores = dataset.mean(axis=1)

labels = []

for score in average_scores:
    if score >= average_scores.mean():
        labels.append("Above Average")
    else:
        labels.append("Below Average")

print("\nStudent Categories:")
print(labels)

y = np.array([1 if label == "Above Average" else 0 for label in labels])

lda = LinearDiscriminantAnalysis(n_components=1)

lda_result = lda.fit_transform(scaled_data, y)

print("\nLDA Result:")
print(lda_result)

dataset["Category"] = labels

print("\nFinal Dataset:")
print(dataset)
```
